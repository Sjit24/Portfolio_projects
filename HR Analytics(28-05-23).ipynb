```python
# Data Preprocessing --
# Null values
# Duplicate records
# Correlation
```

```python
# For checking if null values are present in the data
df.isnull().sum()
```

```python
# If null values were present
# df.dropna() --> For removing the null values
# Replace the null values --> fillna()
#     Data type of the column is object --> replace the null values with mode of the column
#     Data type of the column is int/float --> replace the null values with mean of the column
```

```python
# Checking for duplicate records/rows
# If all the values of a particular row is being repeated in some other row --> duplicate row
df.duplicated().sum()
```

```python
# For removing the duplicate rows -->
df = df.drop_duplicates()
```

```python
# Correlation --> How two columns are related to each other
# Positive --> When the columns are directly proportional to each other
# Negative --> When the columns are inversely proportional to each other
# Zero     --> When the columns are not related

# -1 to 1
df.corr()
```
